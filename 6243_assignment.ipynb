```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
T=2*numpy.pi
dt=0.00001
nt=int(T/dt+1)
t=numpy.linspace(0, T, nt)
Fe=numpy.zeros(nt)
for i in range(1, nt):
    
    Fe[i]=3000+3000*numpy.sin(2*numpy.pi*t[i])
pyplot.plot(t, Fe)
pyplot.xlim(0, 7)
```

```python
Fc=numpy.ones(nt)*1100
Fi=1100
Kc=6
Kb=1
for i in range (1, nt):
    if Fe[i]>=1100*7/6:
        Fc[i]=0
    else:
        Fc[i]=Fi-(Kc/(Kb+Kc))*Fe[i]
Fb=numpy.ones(nt)*1100
for i in range(1, nt):
    if Fe[i]>=1100*7/6:
        Fb[i]=Fe[i]
    else:
        Fb[i]=Fi+(Kb/(Kb+Kc))*Fe[i]
pyplot.plot(t, Fc)
pyplot.plot(t, Fb)
```
